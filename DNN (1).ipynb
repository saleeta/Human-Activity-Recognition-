```python
import pandas as pd 
from pandas import read_csv

def load_file(filepath):
	dataframe = pd.read_excel(filepath, header=0, delimiter=',')
    #dataframe = read_csv(filepath, header=None, delimiter=',')
	return dataframe
 
df = load_file('HAR/Pocket2.xlsx')#Activity_Recognition_DataSet/Pocket.xlsx')
data=df
#data = load_file('HAR/5.csv')
print(data.shape)
```

```python
 
from scipy.spatial.distance import cosine
import numpy as np
from scipy.sparse import csr_matrix
#from __future__ import division
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

mylist = ["Ax", "Ay", "Az","Gx", "Gy", "Gz", "Mx", "My","Mz"]
df.head()
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
def timeSeriesToSupervised(df,var_list,window_size=1, dropnan=True):
    for var in var_list:
        for idx in range(window_size,0,-1):
            var_name = var + '(t+' + str(idx) + ')'    
            df[var_name] = df[var].shift(-idx)

    if dropnan:
        df.dropna(inplace=True)
    return df
```

```python
df
```

```python
df2= timeSeriesToSupervised(df,mylist,100)
```

```python
df2.shape
```

```python

#X = df2.drop('Activity_Label', axis=1)  
X = df2.drop('Time_Stamp', axis=1)  
y = df2['Activity_Label']  
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33) 
```

```python
#check image shape
X_test.shape
X2=X_train
X_test=X_test.drop('Activity_Label', axis=1) 
X_train=X_train.drop('Activity_Label', axis=1) 
X_test.head()
```

```python
#Initializing Neural Network
classifier = Sequential()
```

```python
from keras.utils import to_categorical
y_binary = to_categorical(y_train)
```

```python
X_train.shape
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(activation="softmax", input_dim=909, units=500, kernel_initializer="uniform"))
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=500, kernel_initializer="uniform"))
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=500, kernel_initializer="uniform"))
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=500, kernel_initializer="uniform"))
# Adding the output layer
classifier.add(Dense(output_dim = 7, init = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(X_train, y_binary, batch_size = 500, epochs = 5)
```

```python
# Predicting the Test set results
from sklearn.externals import joblib
y_pred = classifier.predict(X_test)
classifier.save('classifier.h5')
```
