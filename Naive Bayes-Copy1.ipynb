```python
import pandas as pd  
import numpy
import matplotlib.pyplot 
from pandas import read_csv
%matplotlib inline

def load_file(filepath):
	dataframe = pd.read_excel(filepath, header=0, delimiter=',')
    #dataframe = read_csv(filepath, header=None, delimiter=',')
	return dataframe
 
df = load_file('Activity_Recognition_DataSet/Pocket.xlsx')
data=df
#data = load_file('HAR/5.csv')
print(data.shape)
```

```python
def timeSeriesToSupervised(df,var_list,window_size=1, dropnan=True):
    for var in var_list:
        for idx in range(window_size,0,-1):
            var_name = var + '(t+' + str(idx) + ')'    
            df[var_name] = df[var].shift(-idx)

    if dropnan:
        df.dropna(inplace=True)
    return df
```

```python
#ccoppied stuff off o internet


import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

fs = 1000  # Sampling frequency
# Generate the time vector properly
t = np.arange(1000) / fs
signala = np.sin(2*np.pi*100*t) # with frequency of 100
plt.plot(t, signala, label='a')

signalb = np.sin(2*np.pi*20*t) # frequency 20
plt.plot(t, signalb, label='b')

signalc = signala + signalb
plt.plot(t, signalc, label='c')

fc = 30  # Cut-off frequency of the filter
w = fc / (fs / 2) # Normalize the frequency
b, a = signal.butter(5, w, 'low')
output = signal.filtfilt(b, a, signalc)
plt.plot(t, output, label='filtered')
plt.legend()
plt.show()
```

```python
# butterworts filter on stored data

import pandas as pd  
import numpy as np
import matplotlib.pyplot 
from pandas import read_csv
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal



def load_file(filepath):
	dataframe = pd.read_csv(filepath, header=0, delimiter=';')
    #dataframe = read_csv(filepath, header=None, delimiter=',')
	return dataframe

df = load_file('nailbiting_4.csv')
df.head()

```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.plot(df.iloc[:,0],df.iloc[:,1])
fs = 1000
fc = 30  # Cut-off frequency of the filter
w = fc / (fs / 2) # Normalize the frequency
b, a = signal.butter(5, w, 'high')
output = signal.filtfilt(b, a, df.iloc[:,1])
t = df.iloc[:,0]
plt.plot(t, output, label='filtered')
plt.legend()
plt.show()
```
